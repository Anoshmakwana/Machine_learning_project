line plot

## 2. scatter plot

```python
plt.scatter(data['hp'], data['mpg'], marker='*', color='green', label="scatter plot")
plt.xlabel('X value')
plt.ylabel('Y-value')
plt.title("line plot example")
plt.legend()
plt.show()
```

```python
# bar chat
```

```python
category = ['A',  'B', 'C', 'D', 'E']
marks = [15, 24, 30, 20, 5]

# bar chat
plt.bar(category, marks, color='purple', alpha=0.7, label="bar chart")
plt.xlabel('category')
plt.ylabel('marks')
plt.title("bar plot example")
plt.legend()
plt.show()
```

```python
category = ['A',  'B', 'C', 'D', 'E']
marks = [15, 24, 30, 20, 5]

# bar chat
plt.barh(category, marks, color='black', alpha=0.7, label="bar chart")
plt.xlabel('category')
plt.ylabel('marks')
plt.title("bar plot example")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,3))
x = [1,2,3,4,5]
y = [10, 15, 8, 20, 13]
plt.plot(x, y, marker='o', markersize= 10, linestyle='--', color='red', linewidth = 5, label='line plot')

plt.text(2.7,17.5,'learning python', fontsize=12, color='red')
plt.ylabel('Y-value')
plt.title("line plot example")
plt.legend()

plt.savefig('mylineplot.png') # save the plot
plt.show()
```

seaborn library

1. histogram

```python
import seaborn as sns
```

```python
# sample data
data = sns.load_dataset('iris')
```

```python
data.head(3)
```

```python
# create histogram
sns.histplot(data=data, x="sepal_length", kde = True)

# add label and title 
plt.xlabel("sepal length")
plt.ylabel("frequency")
plt.title("histogram of sepal length")

plt.show()
```

```python
# boxplot
```

```python
# sample data
data = sns.load_dataset("tips")
data.head(3)
```

```python
# boxplot
sns.boxplot(data = data, x='day', y="total_bill", palette="Set2")
plt.xlabel("Day")
plt.ylabel("total_bill")
plt.title("box plot of total bill by day")
plt.show()
```

```python
# boxplot
sns.violinplot(data = data, x='day', y="total_bill", palette="pastel")
plt.xlabel("Day")
plt.ylabel("total_bill")
plt.title("violin plot of total bill by day")
plt.show()
```
